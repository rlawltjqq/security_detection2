```python
import cv2
import datetime #날짜 
import time

#def rec():
cap = cv2.VideoCapture(1) # 카메라 1대 사용시 (0)
record = False
if cap.isOpened:
        file_path = "D:\\pass\\test3.mp4"
       
        i=0
        fps = 30
        fourcc = cv2.VideoWriter_fourcc(*'XVID')            # 인코딩 포맷 문자
        width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        size = (int(width), int(height))                   # 프레임 크기
        now = datetime.datetime.now().strftime("%y_%m_%d_%H-%M") #날짜
        record = True
        video = cv2.VideoWriter("D:\\pass\\" + str(now) + ".mp4", fourcc, fps, size)
        
        #out = cv2.VideoWriter(file_path, fourcc, fps, size)  # VideoWriter 객체 생성
        while True:
            ret, frame = cap.read()
            if ret:
                cv2.imshow('camera-recording', frame)               
                
                if record == True: 
                    video.write(frame)
                    print("rec..")
                    
                #for i in range(0,5):
                        
                #        i= i + 1
                #       time.sleep(0.1)
                #cv2.VideoWriter(file_path, fourcc, fps, size)  # 파일 저장
                #for i in range(0,5):
                #        time.sleep(1.0)
                #        cv2.VideoWriter( "%s_%d.mp4" % (file_path,i), fourcc, fps, size)
                #       i= i + 1
                #        time.sleep(0.1)                   
                key = cv2.waitKey(1) & 0xFF
                if key == 27: #esc
                    break
            else:
     
                print('no file!')
                break
        #out.release()                                       # 파일 닫기
else:
        print("Can`t open camera!")
cap.release()
cv2.destroyAllWindows()
```
