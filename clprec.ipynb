```python
import cv2
import datetime #날짜 
import time

cap = cv2.VideoCapture(1)  # 카메라 1대 사용시 (0)
record = False
if cap.isOpened:
        fps = 30
        fourcc = cv2.VideoWriter_fourcc(*'XVID')            # 인코딩 포맷 문자
        width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        size = (int(width), int(height))                   # 프레임 크기
        now = datetime.datetime.now().strftime("%y년 %m월 %d일 %H-%M") #날짜
        record = True
        video = cv2.VideoWriter("D:\\pass\\" + str(now) + ".mp4", fourcc, fps, size) # 파일 경로 / 시간 / 확장자 / 코덱 /fps / size
        
        while True:
            ret, frame = cap.read()
            if ret:
                cv2.imshow('camera-recording', frame)               
                
                if record == True: 
                    video.write(frame)
                 
                    print("rec..")
                                   
                
                key = cv2.waitKey(1) & 0xFF
                if key == 27: #esc
                    break
            else:
     
                print('no file!')
                break
        video.release()                                       # 파일 닫기
else:
        print("Can`t open camera!")
cap.release()
cv2.destroyAllWindows()
```
